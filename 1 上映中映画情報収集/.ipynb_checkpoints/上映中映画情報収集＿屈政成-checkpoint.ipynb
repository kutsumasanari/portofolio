# すべての映画の情報を取得

```python
from selenium.webdriver.chrome import service
from selenium.webdriver.common.by import By
```

```python
titles = []
days = []
countrys = []
scores = []

from selenium.common.exceptions import NoSuchElementException

for page in range(1,14):
    
    url = 'https://filmarks.com/list/now?page={}'.format(page)
    browser.get(url)
        
    elems_wrapper = browser.find_elements_by_class_name('p-content-cassette__title-wrapper')
    for elem_wrapper in elems_wrapper:   
        elem_title = elem_wrapper.find_element_by_class_name('p-content-cassette__title')
        titles.append(elem_title.text)
    
    elems_info = browser.find_elements_by_class_name('p-content-cassette__info__main')
    for elem_info in elems_info:
        elem_day = elem_info.find_element_by_class_name("p-content-cassette__other-info").find_element_by_tag_name("span")
        days.append(elem_day.text)

    elems_info = browser.find_elements_by_class_name('p-content-cassette__other-info')
    for elem_info in elems_info:
        try:
            elem_info.find_element_by_tag_name("a")
        except NoSuchElementException:
            elem_country = "/"
            countrys.append(elem_country)
        else:
            elem_country = elem_info.find_element_by_tag_name("a") 
            countrys.append(elem_country.text)

    elems_star = browser.find_elements_by_class_name('p-content-cassette__rate')     
    for elem_star in elems_star:
        try:
            elem_star.find_element_by_class_name('c-rating__score')
        except NoSuchElementException:
            elem_star = "/"
            scores.append(elem_star)
        else:
            elem_score = elem_star.find_element_by_class_name('c-rating__score')
            scores.append(elem_score.text)
```

```python
df = pd.DataFrame({'映画名': titles,
                   '上映日': days,
                   '製作国': countrys,
                   '評価点数': scores
                  })

df
```

# すべての画像を収集

```python
from PIL import Image
from PIL import UnidentifiedImageError
import io
from urllib import request

for page in range(1,14):
    url = 'https://filmarks.com/list/now?page={}'.format(page)
    browser.get(url)
    elems = browser.find_elements_by_class_name('p-content-cassette__jacket')
    elems_wrapper = browser.find_elements_by_class_name('p-content-cassette__title-wrapper')
    for elem, elem_wrapper in zip(elems, elems_wrapper):
        try:
            elem_title = elem_wrapper.find_element_by_class_name('p-content-cassette__title')
            elem = elem.find_element_by_tag_name('img')
            url = elem.get_attribute('src')
            f = io.BytesIO(request.urlopen(url).read())
            img = Image.open(f)
        except UnidentifiedImageError:
            img = img.convert('RGB')
            img.save('image\{}.jpg'.format(elem_title.text)) 
        else:
            if KeyError:
                img = img.convert('RGB')
                img.save('image\{}.jpg'.format(elem_title.text.replace("/", ' ')))
            else:
                img.save('image\{}.jpg'.format(elem_title.text))
```
