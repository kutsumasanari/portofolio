##  1 データの読み込み

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot 
from sklearn import tree
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot as plt
```

```python
# ファイルの読み込み
train=pd.read_table('./train.tsv')
test=pd.read_table('./test.tsv')
sub=pd.read_csv('./sample_submit.csv',encoding='utf-8', header=None)
```

## 2 前処理（1）

```python
# カテゴリー変数をダミー変数に変換
columns1 = ['cap-shape', 'cap-surface', 'cap-color','bruises','odor','gill-attachment',
            'gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring',
            'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 
            'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color']
X_train=pd.get_dummies(train[columns1])

# 目的変数のYについて、p(毒)の場合は「1」をそれ以外には「0」を入力する。
X_train['Y']=train['Y']
X_train['classes']=X_train['Y'].map(lambda x:1 if x=='p' else 0)
X_train=X_train.drop('Y', axis=1)

# Yを取り出す。
Y=X_train['classes']
X=X_train.drop('classes', axis=1)

# 学習データとテストデータに分ける。
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.05, random_state=1) 
```

## 3 決定木の学習（1）

```python
# モデルの作成
clf=tree.DecisionTreeRegressor(max_depth=10, random_state=1)
clf.fit(X_train, y_train)

#正解率の表示
print ('Score: ',clf.score(X_test, y_test))
```

```python
# 特徴値の重要度の表示。
importances = clf.feature_importances_
indices = np.argsort(importances)

# 重要度の高い順にプロットする。
fig, ax = plt.subplots()
fig.set_size_inches(10, 25)
ax.barh(range(len(importances)), importances[indices])
ax.set_yticks(range(len(importances)))
_ = ax.set_yticklabels(np.array(X_train.columns)[indices])
```

## 2 前処理（2）

```python
# 重要度が高い特徴値だけを残す。
# カテゴリー変数をダミー変数に変換。
columns2 = [ 'odor', 'ring-type', 'cap-color', 'stalk-root', 'stalk-surface-below-ring','spore-print-color']
X_train=pd.get_dummies(train[columns2])

# 目的変数のYについて、p(毒)の場合は「1」をそれ以外には「0」を入力する。
X_train['Y']=train['Y']
X_train['classes']=X_train['Y'].map(lambda x:1 if x=='p' else 0)
X_train=X_train.drop('Y', axis=1)

# Yを取り出す。
Y=X_train['classes']
X=X_train.drop('classes', axis=1)

# 学習データとテストデータに分ける。
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.05, random_state=1) 
```

## 3 決定木の学習（2）

```python
# モデルの作成
clf=tree.DecisionTreeRegressor(max_depth=10, random_state=1)
clf.fit(X_train, y_train)

#正解率の表示
print ('Score: ',clf.score(X_test, y_test))
```

```python
# 特徴値の重要度の表示。
importances = clf.feature_importances_
indices = np.argsort(importances)

# 重要度の高い順にプロットする。
fig, ax = plt.subplots()
fig.set_size_inches(10, 9)
ax.barh(range(len(importances)), importances[indices])
ax.set_yticks(range(len(importances)))
_ = ax.set_yticklabels(np.array(X_train.columns)[indices])
```

## 4 テストデータの予測とデータの導出

```python
# テストデータの予測。
XX_test=pd.get_dummies(test[columns2])
y_pred = clf.predict(XX_test)

# 予測が0.5以上を1、それ以外を0とする。
for i in range(4062):
    if y_pred[i]>=0.5:
        y_pred[i]=1
    else:
        y_pred[i]=0 
df = pd.DataFrame(y_pred, columns = ['Y'])  
```

```python
# idと予測値を統合する。
sub[0] = test['id']
sub[1] = df['Y']

# 予測値の「1」と「0」を「p」と「e」に変換する。
sub[1]=sub[1].replace([1, 0], ["p", "e"])

# 予測データの導出。
sub.to_csv('submission.csv', index=False, header=None)
```
